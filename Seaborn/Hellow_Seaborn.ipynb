# Heatmap

    seaborn.heatmap(data, ​**kwargs)

**cmap**：设置颜色映射（Colormap）。例如：

    sns.heatmap(data, cmap="YlGnBu")  # 使用黄色到蓝色的渐变色

其他常见的 cmap 值包括 "viridis"、"coolwarm"、"RdBu" 等

​**annot**：是否在热力图的每个单元格中显示数值

​**fmt**：当 annot=True 时，设置数值的格式。例如：

    sns.heatmap(data, annot=True, fmt=".2f")  # 显示两位小数

​**cbar**：是否显示颜色条（Colorbar）

```python
plt.Figure(figsize=(20,20))
sns.heatmap(ign_data, cmap='viridis', annot=True, fmt='.2f')
```

## Regression Plot

    seaborn.regplot(x, y, data=None, ​**kwargs)
 ​**fit_reg**：是否拟合回归线。默认为 True。如果设置为 False，则只绘制散点图。

​ **order**：设置多项式回归的阶数。例如，order=2 表示拟合二次多项式回归线。

 **logx**：是否对 x 轴数据进行对数变换。默认为 False。如果设置为 True，则拟合对数回归线。

```python
plt.Figure(figsize=(12,8))
sns.regplot(data=candy_data, x='sugarpercent', y='winpercent')
```

<font size='6'>分面绘图</font>

```python
plt.Figure(figsize=(12,8))
sns.scatterplot(data=candy_data, x='sugarpercent', y='winpercent', hue='chocolate')
```

    seaborn.lmplot(x, y, hue, data, ​**kwargs)

```python
plt.Figure(figsize=(12,8))
sns.lmplot(data=candy_data, x='sugarpercent', y='winpercent', hue='chocolate')
```

    seaborn.swarmplot(x=None, y=None, hue=None, data=None, ​**kwargs)

**orient**：设置蜂群图的方向。可以是 "v"（垂直，默认）或 "h"（水平）。

​**palette**：设置颜色调色板，用于 hue 分组。例如：

    sns.swarmplot(x="category", y="value", hue="group", data=df, palette="Set1")

```python
plt.Figure(figsize=(12,8))
sns.swarmplot(data=candy_data, x='chocolate', y='winpercent', hue='chocolate', palette='Paired')
```

## Palettes

<font size='4'>适用于离散的分类数据，通常用于 hue 参数的分组。</font>

​**deep**：默认调色板，包含深色系。

​**muted**：柔和的颜色，适合一般用途。

​**bright**：鲜艳的颜色，适合强调数据。

​**pastel**：柔和的浅色系，适合背景或次要数据。

​**dark**：深色系，适合对比度高的场景。

​**colorblind**：适合色盲用户，颜色区分明显。

​**Set1** 到 ​**Set3**：预定义的调色板，适合分类数据。

​**Paired**：成对的颜色，适合对比数据。

​**Accent**：强调色，适合突出重点。

<font size='5'>适用于连续数据，颜色从浅到深渐变。</font>

​**Blues**：蓝色渐变。

​**Greens**：绿色渐变。

​**Reds**：红色渐变。

​**Oranges**：橙色渐变。

​**Purples**：紫色渐变。

​**Greys**：灰色渐变。

​**BuGn**：蓝绿色渐变。

​**YlOrRd**：黄橙红色渐变。

```python
plt.Figure(figsize=(12,8))
sns.scatterplot(data=candy_data, x='sugarpercent', y='winpercent', hue='chocolate', palette='Greens')
```

# Set Style

    seaborn.set_style(style=None, rc=None)

Seaborn 提供了以下 5 种预定义样式：

​**darkgrid**​（默认样式）：

    灰色背景 + 白色网格线,适合大多数场景。

​**whitegrid**：

    白色背景 + 灰色网格线,适合强调数据的场景。

​**dark**：

    灰色背景，无网格线,适合简洁的图形。

​**white**：

    白色背景，无网格线,适合简洁的图形。

​**ticks**：

    白色背景，无网格线，但添加了刻度线,适合强调轴刻度的场景。

```python
sns.set_style('whitegrid')
sns.lineplot(spotify_data)
```
