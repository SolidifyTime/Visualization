```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Boxplot

参数 | 作用
|-----|-----|
x / y | 指定横轴和纵轴
data | 数据所在的 DataFrame
hue | 增加分组变量，按颜色显示多个类别
palette | 颜色方案，比如 "Set2"、"pastel" 等
orient | "v"（默认，垂直）或 "h"（水平）
width | 箱子的宽度（默认0.8）

```python
# 加载 seaborn 自带数据集
tips = sns.load_dataset("tips")

# 绘制箱线图（使用所有常用参数）
sns.boxplot(
    x="day",                 # 横轴：星期几
    y="total_bill",          # 纵轴：总账单
    hue="sex",               # 分组：性别
    data=tips,               # 数据源
    palette="Set3",          # 调色板：Set3 风格
    orient="v",              # 垂直方向
    width=0.6,               # 箱子的宽度
    linewidth=1.5,           # 边框线宽
    fliersize=4,             # 异常值点大小
    saturation=0.8           # 颜色饱和度
)
```

# Violin Plot

参数 | 作用
|-----|----|
x / y | 分别设置横纵轴变量
hue | 分组变量
palette | 调色板（颜色方案）
split=True | 让 violin 左右分布显示不同类别（仅限 hue 有两类时）
density_norm | 'area'、'count'、'width' 控制 violin 的宽度
inner | 'box'、'quartile'、'stick'、'point'，控制内部显示
linewidth | 边缘线宽度

```python
sns.violinplot(
    x="day", 
    y="total_bill", 
    hue="sex",               # 按性别分组
    data=tips, 
    palette="Set2",          # 设置调色板
    split=True,              # 将两组 violin 合并成一个图形左右对称展示
    density_norm="count",    # violin 宽度按样本数量缩放
    inner="box",             # 在小提琴内部显示箱线图（可选：box / quartile / point / stick / None）
    linewidth=1.2            # 边线宽度
)
```
